## Calculate Galactic Coordinates

```python
import math

# Zeta Reticuli 		279.0961935830915 -47.2141042298922 39.3

stars = [
    { "name": "vega", "position": [67.5, 19.2, 25]},
    { "name": "alpha centauri", "position": [315.8, -0.7, 4]},
    { "name": "sirius", "position": [227.2, -8.9, 9]},
    { "name": "tau ceti", "position": [173.1, -73.4, 12]},
    { "name": "procyon", "position": [213.7, 13.0, 11]},
    { "name": "40 Eridani", "position": [200.8, -38.0, 16.3]},
    { "name": "Wolf 359", "position": [244.1, 56.1, 7.8]},
    { "name": "Omega Leonis", "position": [223.66, 38.90, 108]},
    { "name": "Canopus", "position": [261.21210208, -25.29220544, 310]},
    { "name": "Capella", "position": [162.58847600, 04.56643053, 42.9]},
    { "name": "Delta Pavonis", "position": [329.7673520576934, -32.4164645155008, 19.89]},
    { "name": "36 Ophiuchi", "position": [358.2788979988, 06.8780379683, 19.5]},
    { "name": "Gliese 581", "position": [354.0776257523032, 40.0186970087399, 20.5]},
    { "name": "Alpha Cassiopeiae", "position": [121.4164062796607, -06.3022507244394, 228]},
    { "name": "Arcturus", "position": [015.05005923, +69.11129459, 36.7]},
    { "name": "Rigel", "position": [209.24118506, -25.24534825, 850]},
    { "name": "Achernar", "position": [290.84124169, -58.79200969, 139]},
    { "name": "Regulus", "position": [226.42735281, +48.93420887, 79]},
    { "name": "Fomalhaut", "position": [020.48814538, -64.90960178, 25]},
    { "name": "Barnard's Star", "position": [031.0087027177046, +14.0626510150033, 5.96]},
    { "name": "Altair", "position": [047.74412569, -08.90918816, 16.7]},
    { "name": "Acrux", "position": [300.12657738, -00.36272479, 321]},
    { "name": "Zeta Reticuli", "position": [279.0961935830915, -47.2141042298922, 39.3]},
]

for star in stars:
    x = star["position"][2] * math.cos(math.radians(star["position"][1])) * math.cos(math.radians(star["position"][0]))
    z = - star["position"][2] * math.cos(math.radians(star["position"][1])) * math.sin(math.radians(star["position"][0]))
    y = - star["position"][2] * math.sin(math.radians(star["position"][1]))
    star["pos_absolute"] = [round(x, 2), round(y, 2), round(z, 2)]

for star in stars:
    print(f"{star['name'].title()}: {star['pos_absolute']}")
```
